# Pre-processing the data

```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', 50)
pd.set_option('display.max_rows', 100)
```

```python
df = pd.read_csv(r'C:/Users/JdMacGyver/Desktop/P_DS/2.DADOS/MICRODADOS_ENADE_2017.txt', sep=';')
```

```python
len(df)
```

### Retirando universidades privadas

```python
df = df[df['CO_CATEGAD']!=4]
```

```python
len(df)
```

### Selecionando alunos do RR

```python
df = df[df['CO_UF_CURSO']==14] # 14 RR
```

```python
len(df)
```

### Columns that we are going to use

```python
used_columns = ['NT_GER']

for i in range(1, 27):
    if(i<10):
        used_columns.append('QE_I0{}'.format(i))
    else:
        used_columns.append('QE_I{}'.format(i))      
```

```python
df = df[used_columns]
```

```python
len(df)
```

### Removing students who does not attending the test

```python
df.dropna(subset=['NT_GER'], axis=0, inplace=True)
```

```python
df['NT_GER'] = df['NT_GER'].apply(lambda x: float(x.replace(',','.')))
```

```python
df.head(5)
```

### Categorizacao das notas

```python
def cat_notas(df):
    if df['NT_GER'] < 25.0:
        return '0_25'
    elif df['NT_GER'] >= 25.0 and df['NT_GER'] <= 49.9:
        return '25_50'
    elif df['NT_GER'] >= 50.0 and df['NT_GER'] <= 74.9:
        return '50_75'
    else:
        return '75_100'
```

```python
df['NT_GER'] = df.apply(cat_notas, axis=1)
```

```python
df.dropna(how='any', inplace=True)
```

```python
len(df)
```

```python
df.head(5)
```

```python
df['QE_I16'] = df['QE_I16'].astype(int)
```

```python
df['QE_I16'] = df['QE_I16'].apply(str)
```

```python
df.head(5)
```

```python
colunas = [col for col in df.columns if col != 'NT_GER']
```

```python
for coluna in colunas:
    for row in df.index:
        df[coluna][row] = (coluna.replace('E_I','')) + df[coluna][row]
```

```python
df
```

```python
df['QE_I16'] = df['QE_I16'].apply(lambda x: x.replace('Q16', 'Q16_'))
```

```python
df.head(5)
```

```python
df.to_csv('RR.csv', index=False)
```
